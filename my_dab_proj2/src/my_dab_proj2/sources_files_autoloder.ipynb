```python
%sql
CREATE VOLUME IF NOT EXISTS etl_pipeline_catalog.etl_pipeline_schema.etl_cloudsource_vol;
CREATE VOLUME IF NOT EXISTS etl_pipeline_catalog.etl_pipeline_schema.etl_bronzezone_vol
```

```python
from pyspark.sql.functions import lit,col
staff1_df=spark.readStream.format("cloudFiles")\
.option("cloudFiles.format","csv")\
.option("cloudFiles.inferColumnTypes",True)\
.option("cloudFiles.maxFilesPerTrigger",1)\
.option("schemaEvolutionMode","addnewColumns")\
.option("cloudFiles.maxFilesPerTrigger",1)\
.option("header",True)\
.option("cloudFiles.schemaLocation","/Volumes/etl_pipeline_catalog/etl_pipeline_schema/etl_bronzezone_vol/schema/staff1/")\
.option("checkPointLocation","/Volumes/etl_pipeline_catalog/etl_pipeline_schema/etl_bronzezone_vol/chkpoint/staff1/")\
.load("/Volumes/etl_pipeline_catalog/etl_pipeline_schema/etl_cloudsource_vol/staff1/")

staff2_df=spark.readStream.format("cloudFiles")\
.option("cloudFiles.format","csv")\
.option("cloudFiles.inferColumnTypes",True)\
.option("cloudFiles.maxFilesPerTrigger",1)\
.option("schemaEvolutionMode","addnewColumns")\
.option("header",True)\
.option("cloudFiles.schemaLocation","/Volumes/etl_pipeline_catalog/etl_pipeline_schema/etl_bronzezone_vol/schema/staff2/")\
.option("checkPointLocation","/Volumes/etl_pipeline_catalog/etl_pipeline_schema/etl_bronzezone_vol/chkpoint/staff2/")\
.load("/Volumes/etl_pipeline_catalog/etl_pipeline_schema/etl_cloudsource_vol/staff2/")

shipment_df=spark.readStream.format("cloudFiles")\
.option("cloudFiles.format","json")\
.option("multiLine",True)\
.option("cloudFiles.maxFilesPerTrigger",1)\
.option("cloudFiles.schemaLocation","/Volumes/etl_pipeline_catalog/etl_pipeline_schema/etl_bronzezone_vol/schema/shipment_details/")\
.option("checkPointLocation","/Volumes/etl_pipeline_catalog/etl_pipeline_schema/etl_bronzezone_vol/chkpoint/shipment_details/")\
.load("/Volumes/etl_pipeline_catalog/etl_pipeline_schema/etl_cloudsource_vol/shipment_details/")

master_city=spark.readStream.format("cloudFiles")\
.option("cloudFiles.format","csv")\
.option("cloudFiles.inferColumnTypes",True)\
.option("schemaEvolutionMode","addnewColumns")\
.option("cloudFiles.maxFilesPerTrigger",1)\
.option("header",True)\
.option("cloudFiles.schemaLocation","/Volumes/etl_pipeline_catalog/etl_pipeline_schema/etl_bronzezone_vol/schema/master_city/")\
.option("checkPointLocation","/Volumes/etl_pipeline_catalog/etl_pipeline_schema/etl_bronzezone_vol/chkpoint/master_city/")\
.load("/Volumes/etl_pipeline_catalog/etl_pipeline_schema/etl_cloudsource_vol/master_city/")
```

```python
staff1_df.writeStream.trigger(availableNow=True).format("delta")\
.option("mergeSchema",True)\
.option("checkPointLocation","/Volumes/etl_pipeline_catalog/etl_pipeline_schema/etl_bronzezone_vol/chkpoint/staff1/")\
.start("/Volumes/etl_pipeline_catalog/etl_pipeline_schema/etl_bronzezone_vol/staff1/")

staff2_df.writeStream.trigger(availableNow=True).format("delta")\
.option("mergeSchema",True)\
.option("checkPointLocation","/Volumes/etl_pipeline_catalog/etl_pipeline_schema/etl_bronzezone_vol/chkpoint/staff2/")\
.start("/Volumes/etl_pipeline_catalog/etl_pipeline_schema/etl_bronzezone_vol/staff2/")

shipment_df.writeStream.trigger(availableNow=True).format("delta")\
.option("checkPointLocation","/Volumes/etl_pipeline_catalog/etl_pipeline_schema/etl_bronzezone_vol/chkpoint/shipment_details/")\
.start("/Volumes/etl_pipeline_catalog/etl_pipeline_schema/etl_bronzezone_vol/shipment_details/")

master_city.writeStream.trigger(availableNow=True).format("delta")\
.option("mergeSchema",True)\
.option("checkPointLocation","/Volumes/etl_pipeline_catalog/etl_pipeline_schema/etl_bronzezone_vol/chkpoint/master_city/")\
.start("/Volumes/etl_pipeline_catalog/etl_pipeline_schema/etl_bronzezone_vol/master_city/")
```
